```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv('dataset.csv')
data = data.dropna()
```

```python
pd.DataFrame(data)
```

```python
data_train = data.iloc[:int(len(data)*0.8)]
data_train
#accepts 2 arguments
```

```python
data_test = data.iloc[int(0.8*len(data)):]
data_test
```

```python
#training data
x = np.array(data_train.Humidity)
y = np.array(data_train.Temperature)
```

```python
def plot(x,y):
    plt.plot(x,y,'o')
    plt.xlabel("Humdity")
    plt.ylabel("Temperature")
```

```python
plot(x,y)
```

```python
#hypothesis
theta = np.random.rand(2,1)
np.random.seed(2)

def hypothesis(X, theta):
    return theta[0] + theta[1] * X
```

```python
def line(theta):
    axes = plt.gca()
    x_value = np.array(axes.get_xlim())
    y_value = hypothesis(x_value, theta)
    plt.plot(x_value, y_value, '--')
```

```python
def draw(x, y, theta):
    plot(x,y)
    line(theta)
```

```python
draw(x, y, theta)
```

```python
m = len(x)
def cost(x, y, theta):
    return (1/(2*m)) * np.sum((hypothesis(x, theta)-y)**2)
```

```python
cost(x, y, theta)
```

```python
def gradient_descent(x, y, theta, alpha, steps):
    for i in range(steps):
        old_cost = cost(x, y, theta)
        temp_0 = theta[0] - alpha * ((1/m) * np.sum(hypothesis(x, theta) - y))
        temp_1 = theta[1] - alpha * ((1/m) *np.dot((hypothesis(x, theta) - y), x))
        theta[0], theta[1] = temp_0, temp_1
        new_cost = cost(x, y, theta)
        if new_cost < old_cost:
            print("Decreaing: ", new_cost)
        else:
            print("INCREASE")
    
```

```python
gradient_descent(x, y, theta, alpha=0.01, steps=10000)
theta
```

```python
draw(x, y, theta)
```

```python
test_x = np.array(data_test.Humidity)
test_y = np.array(data_test.Temperature)
```

```python
draw(test_x, test_y, theta)
```

```python
def error(testx, testy, theta):
    test_predictiony = hypothesis(testx, theta)
    return abs(testy - test_predictiony)
```

```python
m = len(test_x)
s = 0
for i in range(m):
    s += error(test_x[i], test_y[i], theta)
s/m
```
